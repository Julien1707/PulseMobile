```python
import pandas as pd
import matplotlib.pyplot as plt

# Charger le fichier
file_path = 'base_données/Best Selling Mobile Phones 2020.csv'
data = pd.read_csv(file_path)

# Diagramme en barres des ventes par modèle de téléphone
plt.figure(figsize=(10, 6))
plt.bar(data['Phone'], data['Sold(million)'], color='skyblue')
plt.xlabel('Modèle de Téléphone')
plt.ylabel('Ventes (millions)')
plt.title('Ventes par Modèle de Téléphone en 2020')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Diagramme circulaire pour les parts de marché par entreprise
company_sales = data.groupby('Company')['Sold(million)'].sum()

plt.figure(figsize=(8, 8))
plt.pie(company_sales, labels=company_sales.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title('Part de Marché par Entreprise en 2020')
plt.show()
```

```python
# Utilisation de K-means
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd
import matplotlib.pyplot as plt

# Charger les données
file_path = 'base_données//bsp.csv'  # Remplace par le chemin de ton fichier
data = pd.read_csv(file_path)

# Garder seulement la colonne numérique
data_numeric = data[['units_sold_m']]

# Standardisation des données
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_numeric)

# Appliquer K-means avec 3 clusters
kmeans = KMeans(n_clusters=3, random_state=0)
data['Cluster'] = kmeans.fit_predict(data_scaled)

# Afficher le résultat
print(data[['manufacturer', 'model', 'form', 'smartphone', 'year', 'units_sold_m', 'Cluster']])




# 1. Diagramme en barres des unités vendues par fabricant
manufacturer_sales = data.groupby('manufacturer')['units_sold_m'].sum().sort_values()

plt.figure(figsize=(10, 6))
manufacturer_sales.plot(kind='barh', color='skyblue')
plt.xlabel('Unité vendues (en millions)')
plt.ylabel('Fabricant')
plt.title('Total des ventes par fabricant')
plt.tight_layout()
plt.show()

# 2. Nuage de points des ventes par année, colorié par cluster
plt.figure(figsize=(10, 6))
for cluster in data['Cluster'].unique():
    cluster_data = data[data['Cluster'] == cluster]
    plt.scatter(cluster_data['year'], cluster_data['units_sold_m'], label=f'Cluster {cluster}')

plt.xlabel('Année')
plt.ylabel('Unité vendues (en millions)')
plt.title('Répartition des ventes par année et par cluster')
plt.legend()
plt.show()

# Filtrer les produits du Cluster 1
cluster_1_products = data[data['Cluster'] == 1][['manufacturer', 'model', 'units_sold_m']]

# Afficher les résultats pour Cluster 1
print(cluster_1_products)


```

```python
import pandas as pd
import os

# Dossier contenant les fichiers CSV
dossier_csv = 'base_données'  # Remplace par le chemin de ton dossier

# Lister tous les fichiers CSV du dossier
fichiers_csv = [f for f in os.listdir(dossier_csv) if f.endswith('.csv')]

# Parcourir chaque fichier CSV
for fichier in fichiers_csv:
    chemin_fichier = os.path.join(dossier_csv, fichier)
    print(f"\n--- Traitement du fichier : {fichier} ---")
    
    # Charger le fichier CSV dans un DataFrame
    try:
        df = pd.read_csv(chemin_fichier)
    except Exception as e:
        print(f"Erreur lors du chargement du fichier {fichier} : {e}")
        continue  # Passer au fichier suivant s'il y a une erreur
    
    # Supprimer les lignes entièrement vides
    df = df.dropna(how='all')
    
    # Supprimer les doublons
    df = df.drop_duplicates()

    # Vérifier et afficher les lignes restantes avec des valeurs manquantes
    lignes_manquantes = df[df.isnull().any(axis=1)]
    
    if not lignes_manquantes.empty:
        print(f"Lignes avec des valeurs manquantes dans {fichier} :")
        print(lignes_manquantes)
    else:
        print(f"Aucune ligne avec des valeurs manquantes dans {fichier}.")

    # Sauvegarder le DataFrame nettoyé dans le même fichier
    df.to_csv(chemin_fichier, index=False)
    print(f"Fichier {fichier} mis à jour avec les lignes vides et doublons supprimés.")

    
```
